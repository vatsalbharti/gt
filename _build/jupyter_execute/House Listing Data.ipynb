***99 acres , magic bricks and housing.com are real estate platforms where one can find house,commerial,appartments,flats,lands etc. listing , we can use such data points to understand how rental and buy prices are indicators of affluence and socioeconomic status in a given area***.

`Income`: Generally, higher-income individuals and families can afford to pay more for housing, whether it's rent or a mortgage. Therefore, areas with higher rental and purchase prices often have a higher average income. Conversely, areas with lower rental and purchase prices tend to have a lower average income.

`Wealth`: Home ownership, in particular, is a key indicator of wealth. The more expensive homes are in an area, the wealthier the residents of that area are likely to be.

`Demand`: High rental and purchase prices can also indicate high demand for housing in an area. This could be because the area has desirable amenities, good schools, low crime rates, or other factors that attract affluent residents.

`Quality of Life`: Higher housing costs are often associated with higher quality of life, including access to good schools, parks, recreational facilities, and other community resources. These factors are typically more available or of higher quality in more affluent areas.

`Education and Occupation`: Areas with higher housing costs often have residents with higher levels of education and more prestigious occupations. This is because income (which is correlated with education and occupation) is a major factor in the ability to afford housing.

However, while rental and purchase prices can be good indicators of affluence, they are not perfect. They do not account for wealth that is not tied up in real estate, and they do not consider income inequality within an area. Furthermore, in some cases, individuals or families may choose to live in a less expensive area to save money or may be living in an expensive area due to factors other than their personal wealth (for example, being supported by family or receiving housing subsidies). It acts as a great proxy data for the same

```python
rent_data['LOCALITY'] = rent_data['LOCALITY'] + "," +rent_data['City based on hdfc address']

# Replace 'Price on Request' with NaN in PRICE column
rent_data['PRICE'] = pd.to_numeric(rent_data['PRICE'].str.replace(',', ''), errors='coerce')

# Replace non-numerical values with NaN in AREA column
rent_data['AREA'] = pd.to_numeric(rent_data['AREA'].str.replace(' sq.ft.', ''), errors='coerce')

# Calculate the average rental price per square foot for each city
city_avg_price = rent_data.groupby('City based on hdfc address')['PRICE'].mean().sort_values(ascending=False)
top_cities = city_avg_price.head(10)


# Calculate the average rental price per square foot for each locality
locality_avg_price = rent_data.groupby('LOCALITY')['PRICE'].mean().sort_values(ascending=False)

# Display the top 10 localities with the highest average rental price
top_localities = locality_avg_price.head(10)

import matplotlib.pyplot as plt

# Create bar plots
fig, axs = plt.subplots(2, 1, figsize=(14, 10))

# Bar plot for top cities
axs[0].bar(top_cities.index, top_cities.values, color='b')
axs[0].set_title('Top 10 Cities by Average Rental Price')
axs[0].set_xlabel('City')
axs[0].set_ylabel('Average Rental Price')
plt.setp(axs[0].xaxis.get_majorticklabels(), rotation=45)

# Bar plot for top localities
axs[1].bar(top_localities.index, top_localities.values, color='g')
axs[1].set_title('Top 10 Localities by Average Rental Price')
axs[1].set_xlabel('Locality')
axs[1].set_ylabel('Average Rental Price')
plt.setp(axs[1].xaxis.get_majorticklabels(), rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

<br>

`City-level price type analysis`: Analyze the distribution of property types within each city.

<br>

```python
# City-level property type analysis
city_property_type_counts = rent_data.groupby(['City based on hdfc address', 'PROPERTY_TYPE']).size().unstack().fillna(0)

# Filter data for top cities
top_cities_data = rent_data[rent_data['City based on hdfc address'].isin(top_cities.index)]

# Create boxplots for rental prices in top cities
plt.figure(figsize=(14, 6))
boxplot = top_cities_data.boxplot(column='PRICE', by='City based on hdfc address', grid=False, vert=False, showfliers=False)
plt.title('City-Level Rental Price Distribution for Top 10 Cities')
plt.suptitle('') # Remove automatically generated title
plt.xlabel('Price')
plt.ylabel('City')
plt.show()
```

<br>

The boxplot above provides a summary of the property sizes (in terms of area) distribution in the top 10 cities. The line inside each box represents the median area in the city, while the box itself represents the interquartile range (IQR, or the middle 50% of areas). Again, extreme outliers have been omitted from the plot for readability.

```python
# Calculate the average rental price per square foot for each city
city_avg_price = rent_data.groupby('CITY')['PRICE'].mean().sort_values(ascending=False)
top_cities = city_avg_price.head(20)

# Filter data for top cities
top_cities_data = rent_data[rent_data['CITY'].isin(top_cities.index)]

# Create boxplots for rental prices in top cities
plt.figure(figsize=(14, 6))
boxplot = top_cities_data.boxplot(column='PRICE', by='CITY', grid=False, vert=False, showfliers=False)
plt.title('City-Level Rental Price Distribution for Top 20 localitywise in Cities')
plt.suptitle('') # Remove automatically generated title
plt.xlabel('Price')
plt.ylabel('City')
plt.show()
```

```python
# Create boxplots for area in top cities
plt.figure(figsize=(14, 6))
boxplot = top_cities_data.boxplot(column='AREA', by='CITY', grid=False, vert=False, showfliers=False)
plt.title('City-Level Area Distribution for Top 20 localitywise in Cities')
plt.suptitle('') # Remove automatically generated title
plt.xlabel('Area (sq.ft.)')
plt.ylabel('City')
plt.show()
```

```python
# Set up the figure and axes for the histograms
fig, axs = plt.subplots(len(top_cities.index), 1, figsize=(10, 20))

# Generate a histogram for each city
for ax, city in zip(axs, top_cities.index):
    city_data = top_cities_data[top_cities_data['CITY'] == city]
    ax.hist(city_data['AREA'], bins=50, color='skyblue', edgecolor='black')
    ax.set_title(city)
    ax.set_xlim([0, 5000])  # Limit x-axis for readability
    ax.set_xlabel('Area (sq.ft.)')
    ax.set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

<br>

The histograms above show the distribution of property sizes (in terms of area) within each of the top 10 cities. Each histogram represents a different city, and the x-axis represents the area of properties. The y-axis represents the count of properties within each area range (bin). For readability, the x-axis is limited to properties with areas less than 5000 sq.ft. Properties larger than this are considered outliers for the purpose of this visualization.

We can see that for most cities, the majority of properties are less than 2000 sq.ft. in size, with very few properties larger than this. However, the distribution varies from city to city.
